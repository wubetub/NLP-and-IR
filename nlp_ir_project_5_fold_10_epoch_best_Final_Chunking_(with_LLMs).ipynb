```python
# === ✅ Part 1: Imports, Setup, Normalization, Dataset Loading ===

import os
import re
import gc
import joblib
import shutil
import warnings
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import gradio as gr
import torch
import torch.nn as nn

from sklearn.model_selection import StratifiedKFold, train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.metrics import (
    classification_report, confusion_matrix, ConfusionMatrixDisplay,
    precision_recall_fscore_support, accuracy_score, f1_score, recall_score, precision_score
)
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.utils.class_weight import compute_class_weight
from scipy.stats import ttest_rel, wilcoxon

from gensim.models import Word2Vec
from gensim.utils import simple_preprocess

from transformers import (
    AutoTokenizer, AutoModelForSequenceClassification,
    Trainer, TrainingArguments, EarlyStoppingCallback
)

warnings.filterwarnings("ignore")
os.environ["TRANSFORMERS_NO_SAFE_SERIALIZATION"] = "1"
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("🖥️ Device in use:", device)

# === ✅ Amharic Normalization Function ===
def preprocess_amharic(text):
    text = str(text)
    patterns = [
        ('[ሃሕላሐሓዻ]', 'ሀ'), ('[ሑሁዅ]', 'ሁ'), ('[ሒሒዺ]', 'ሂ'), ('[ሔሔዌ]', 'ሄ'),
        ('[ህሕ]', 'ህ'), ('[ሖሖዾ]', 'ሆ'), ('[ሠ]', 'ሰ'), ('[ሡ]', 'ሱ'), ('[ሢ]', 'ሲ'),
        ('[ሣ]', 'ሳ'), ('[ሤ]', 'ሴ'), ('[ሥ]', 'ስ'), ('[ሦ]', 'ሶ'), ('[ዓሣዐ]', 'አ'),
        ('[ዑ]', 'ኡ'), ('[ዒ]', 'ኢ'), ('[ዔ]', 'ኤ'), ('[ዕ]', 'እ'), ('[ዖ]', 'ኦ'),

    ]
    for pattern, rep in patterns:
        text = re.sub(pattern, rep, text)
    text = re.sub(r"[^\u1200-\u137F\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

# === ✅ Load and Preprocess Dataset ===
df1 = pd.read_excel("/content/dataset_1.xlsx", engine="openpyxl")
df2 = pd.read_excel("/content/dataset_2.xlsx", engine="openpyxl")

# Raw Data Overview (before preprocessing)
raw_df = pd.concat([df1, df2], ignore_index=True)
raw_df.dropna(subset=["text", "label"], inplace=True)
raw_df["label"] = raw_df["label"].astype(str)
raw_counts = raw_df["label"].value_counts().sort_index()
print("\n📄 Raw Label Counts:\n", raw_counts)

# Normalize and retain all rows
df = raw_df.copy()
df["text"] = df["text"].astype(str).apply(preprocess_amharic)

# Encode labels
df["label"] = pd.Categorical(df["label"])
label_names = list(df["label"].cat.categories)
label2id = {label: i for i, label in enumerate(label_names)}
id2label = {i: label for label, i in label2id.items()}
df["label_id"] = df["label"].apply(lambda x: label2id[x])
joblib.dump(id2label, "label_map.pkl")

# Word length column for analysis
df["word_len"] = df["text"].str.split().apply(len)

# Shuffle and deduplicate
df = df.drop_duplicates(subset=["text", "label"]).sample(frac=1, random_state=42).reset_index(drop=True)

# === ✅ Class Distribution Summary After Preprocessing ===
print("\n📊 Sample Data:\n", df[["text", "label", "word_len"]].sample(20))
label_counts = df["label"].value_counts().sort_index()
label_props = df["label"].value_counts(normalize=True).sort_index()
label_df = pd.DataFrame({
    "Label": label_counts.index,
    "Count": label_counts.values,
    "Proportion": label_props.values
})
print("\n📈 Class Distribution After Preprocessing:\n", label_df.to_string(index=False))

# === ✅ Plot Class Distribution ===
plt.figure(figsize=(10, 6))
sns.barplot(x=label_counts.index, y=label_counts.values)
plt.title("Class Distribution After Preprocessing")
plt.xlabel("Label")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid()
plt.tight_layout()
plt.show()

# === ✅ Add for per-class metric aggregation ===
all_per_class_dfs = []


# === ✅ Part 2: Dataset Split, Class Weights, Dataset Wrapper, Trainer Setup ===

# === ✅ Train/Test Split ===
df_trainval, df_test = train_test_split(
    df, test_size=0.1, stratify=df["label_id"], random_state=42
)
df_test["text"] = df_test["text"].astype(str).apply(preprocess_amharic)

# === ✅ Compute Class Weights (used in CrossEntropyLoss) ===
weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(df_trainval["label_id"]),
    y=df_trainval["label_id"]
)
class_weights = torch.tensor(weights, dtype=torch.float).to(device)

# === ✅ Dataset Format Class for Transformers ===
class NewsDataset(torch.utils.data.Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __getitem__(self, idx):
        item = {k: torch.tensor(v[idx]) for k, v in self.encodings.items()}
        item["labels"] = torch.tensor(self.labels[idx])
        return item

    def __len__(self):
        return len(self.labels)

# === ✅ Enhanced Trainer for Weighted Loss & Metric Tracking ===
class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False):
        labels = inputs.pop("labels")
        outputs = model(**inputs)
        logits = outputs.logits
        loss_fct = nn.CrossEntropyLoss(weight=class_weights)
        loss = loss_fct(logits, labels)
        return (loss, outputs) if return_outputs else loss


# === ✅ Part 3: Transformer Training with Logging and Metric Tracking ===

skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

model_paths = {
    "afriberta": "castorini/afriberta_base",
    "afroxlmr": "Davlan/afro-xlmr-base"
}

results = []
loss_logs = []
conf_matrices = {key: [] for key in model_paths}
epoch_logs = []

for model_key, model_path in model_paths.items():
    tokenizer = AutoTokenizer.from_pretrained(model_path)
    print(f"\n🚀 Training Transformer Model: {model_key.upper()}")

    for fold, (train_idx, val_idx) in enumerate(skf.split(df_trainval["text"], df_trainval["label_id"])):
        print(f"\n📂 Fold {fold + 1}/5")

        train_texts = df_trainval.iloc[train_idx]["text"].tolist()
        val_texts = df_trainval.iloc[val_idx]["text"].tolist()
        train_labels = df_trainval.iloc[train_idx]["label_id"].tolist()
        val_labels = df_trainval.iloc[val_idx]["label_id"].tolist()

        train_enc = tokenizer(train_texts, truncation=True, padding="max_length", max_length=512)
        val_enc = tokenizer(val_texts, truncation=True, padding="max_length", max_length=512)

        train_dataset = NewsDataset(train_enc, train_labels)
        val_dataset = NewsDataset(val_enc, val_labels)

        model = AutoModelForSequenceClassification.from_pretrained(model_path, num_labels=len(label_names)).to(device)

        args = TrainingArguments(
            output_dir=f"./{model_key}_fold{fold + 1}",
            evaluation_strategy="epoch",
            save_strategy="epoch",
            logging_strategy="epoch",
            save_total_limit=1,
            num_train_epochs=10,
            per_device_train_batch_size=8,
            per_device_eval_batch_size=8,
            learning_rate=2e-5,
            warmup_ratio=0.1,
            weight_decay=0.01,
            report_to="none",
            load_best_model_at_end=True,
            metric_for_best_model="eval_macro_f1",
            greater_is_better=True
        )

        def compute_metrics(eval_pred):
            preds = np.argmax(eval_pred.predictions, axis=1)
            prec, rec, f1, _ = precision_recall_fscore_support(
                eval_pred.label_ids, preds, average="weighted", zero_division=0)
            macro_f1 = f1_score(eval_pred.label_ids, preds, average="macro", zero_division=0)
            acc = accuracy_score(eval_pred.label_ids, preds)
            return {
                "accuracy": acc,
                "precision": prec,
                "recall": rec,
                "f1": f1,
                "macro_f1": macro_f1
            }

        trainer = WeightedTrainer(
            model=model,
            args=args,
            train_dataset=train_dataset,
            eval_dataset=val_dataset,
            compute_metrics=compute_metrics,
            callbacks=[EarlyStoppingCallback(early_stopping_patience=2)]
        )

        trainer.train()
        trainer.save_model(f"./final_model_{model_key}_fold{fold + 1}")

        # Save logs per fold/model
        logs = pd.DataFrame(trainer.state.log_history)
        logs["Model"] = model_key
        logs["Fold"] = fold + 1
        loss_logs.append(logs)

        # Per-epoch metrics (loss/accuracy/precision/recall/f1/macro-F1)
        for entry in trainer.state.log_history:
            if "epoch" in entry:
                epoch_logs.append({
                    "Model": model_key,
                    "Fold": fold + 1,
                    "Epoch": entry.get("epoch"),
                    "Train_Loss": entry.get("loss", np.nan),
                    "Eval_Loss": entry.get("eval_loss", np.nan),
                    "Accuracy": entry.get("eval_accuracy", np.nan),
                    "Precision": entry.get("eval_precision", np.nan),
                    "Recall": entry.get("eval_recall", np.nan),
                    "F1": entry.get("eval_f1", np.nan),
                    "Macro_F1": entry.get("eval_macro_f1", np.nan)
                })

        # Evaluation
        eval_metrics = trainer.evaluate()
        preds = trainer.predict(val_dataset)
        y_true = preds.label_ids
        y_pred = np.argmax(preds.predictions, axis=1)

        cm = confusion_matrix(y_true, y_pred)
        conf_matrices[model_key].append(cm)

        results.append({
            "Model": model_key,
            "Fold": fold + 1,
            "Accuracy": eval_metrics["eval_accuracy"],
            "Precision": eval_metrics["eval_precision"],
            "Recall": eval_metrics["eval_recall"],
            "F1": eval_metrics["eval_f1"],
            "Macro_F1": eval_metrics.get("eval_macro_f1", 0)
        })

        # ✅ Collect Per-Class Metrics
        report_dict = classification_report(y_true, y_pred, target_names=label_names, output_dict=True, zero_division=0)
        report_df = pd.DataFrame(report_dict).transpose()
        report_df["Model"] = model_key
        report_df["Fold"] = fold + 1
        report_df.reset_index(inplace=True)
        report_df.rename(columns={"index": "Class"}, inplace=True)
        all_per_class_dfs.append(report_df)

        print(f"\n📊 Fold {fold + 1} Report ({model_key.upper()}):")
        print(classification_report(y_true, y_pred, target_names=label_names, zero_division=0))

        del model
        torch.cuda.empty_cache()
        gc.collect()

# === ✅ Part 4: Plot Training vs. Validation Curves (Loss, Accuracy) ===

# Convert epoch logs to DataFrame
epoch_df = pd.DataFrame(epoch_logs)

# Drop rows where both train and eval loss are NaN
epoch_df.dropna(subset=["Train_Loss", "Eval_Loss", "Accuracy"], how="all", inplace=True)

# === ✅ Plotting Functions ===
def plot_loss_curves(df, model_key):
    fig, ax = plt.subplots(figsize=(7, 4))
    for fold in sorted(df["Fold"].unique()):
        fold_df = df[df["Fold"] == fold]
        ax.plot(fold_df["Epoch"], fold_df["Train_Loss"], label=f"Train Loss (Fold {fold})", linestyle="--")
        ax.plot(fold_df["Epoch"], fold_df["Eval_Loss"], label=f"Val Loss (Fold {fold})")
    ax.set_title(f"Training vs Validation Loss: {model_key.upper()}")
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Loss")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    plt.show()

def plot_accuracy_curves(df, model_key):
    fig, ax = plt.subplots(figsize=(7, 4))
    for fold in sorted(df["Fold"].unique()):
        fold_df = df[df["Fold"] == fold]
        ax.plot(fold_df["Epoch"], fold_df["Accuracy"], label=f"Val Acc (Fold {fold})")
    ax.set_title(f"Validation Accuracy: {model_key.upper()}")
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Accuracy")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    plt.show()

# === ✅ Plot Curves for Each Transformer Model ===
for model_key in model_paths:
    model_df = epoch_df[epoch_df["Model"] == model_key]
    if not model_df.empty:
        plot_loss_curves(model_df, model_key)
        plot_accuracy_curves(model_df, model_key)


# === ✅ Part 4 (Updated): Plot Training vs. Validation Curves (Loss, Accuracy) ===

# Convert epoch logs to DataFrame
epoch_df = pd.DataFrame(epoch_logs)
epoch_df.dropna(subset=["Train_Loss", "Eval_Loss", "Accuracy"], how="all", inplace=True)

# === ✅ Directory to Save Plots ===
os.makedirs("plots", exist_ok=True)

# === ✅ Plot Per-Fold Loss Curves ===
def plot_loss_curves(df, model_key):
    fig, ax = plt.subplots(figsize=(7, 4))
    for fold in sorted(df["Fold"].unique()):
        fold_df = df[df["Fold"] == fold]
        ax.plot(fold_df["Epoch"], fold_df["Train_Loss"], label=f"Train Loss (Fold {fold})", linestyle="--")
        ax.plot(fold_df["Epoch"], fold_df["Eval_Loss"], label=f"Val Loss (Fold {fold})")
    ax.set_title(f"{model_key.upper()} - Train vs Validation Loss (Per Fold)")
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Loss")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    filename = f"plots/{model_key}_loss_per_fold.png"
    plt.savefig(filename)
    plt.show()

# === ✅ Plot Per-Fold Accuracy Curves ===
def plot_accuracy_curves(df, model_key):
    fig, ax = plt.subplots(figsize=(7, 4))
    for fold in sorted(df["Fold"].unique()):
        fold_df = df[df["Fold"] == fold]
        ax.plot(fold_df["Epoch"], fold_df["Accuracy"], label=f"Val Acc (Fold {fold})", marker='o')
    ax.set_title(f"{model_key.upper()} - Validation Accuracy (Per Fold)")
    ax.set_xlabel("Epoch")
    ax.set_ylabel("Accuracy")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    filename = f"plots/{model_key}_val_acc_per_fold.png"
    plt.savefig(filename)
    plt.show()

# === ✅ Plot Average Loss and Accuracy Across Folds ===
def plot_avg_loss_accuracy(df, model_key):
    model_df = df[df["Model"] == model_key]
    if model_df.empty:
        return
    grouped = model_df.groupby("Epoch").agg({
        "Train_Loss": "mean",
        "Eval_Loss": "mean",
        "Accuracy": "mean"
    }).reset_index()

    # Avg Loss Curve
    plt.figure(figsize=(7, 4))
    plt.plot(grouped["Epoch"], grouped["Train_Loss"], label="Avg Train Loss", linestyle="--")
    plt.plot(grouped["Epoch"], grouped["Eval_Loss"], label="Avg Val Loss")
    plt.title(f"{model_key.upper()} - Avg Train vs Val Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"plots/{model_key}_avg_loss_curve.png")
    plt.show()

    # Avg Accuracy Curve
    plt.figure(figsize=(7, 4))
    plt.plot(grouped["Epoch"], grouped["Accuracy"], marker='o', label="Avg Val Accuracy")
    plt.title(f"{model_key.upper()} - Avg Validation Accuracy")
    plt.xlabel("Epoch")
    plt.ylabel("Accuracy")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"plots/{model_key}_avg_accuracy_curve.png")
    plt.show()

# === ✅ Plot + Save Curves Per Model ===
for model_key in model_paths.keys():
    model_df = epoch_df[epoch_df["Model"] == model_key]
    if not model_df.empty:
        # Per-fold plots
        plot_loss_curves(model_df, model_key)
        plot_accuracy_curves(model_df, model_key)

        # Average across folds
        plot_avg_loss_accuracy(model_df, model_key)


# === ✅ Part 5: Holdout Test Evaluation for Transformer Models ===

for model_key, model_path in model_paths.items():
    print(f"\n🧪 Evaluating {model_key.upper()} on Hold-out Test Set")

    tokenizer = AutoTokenizer.from_pretrained(model_path)
    test_enc = tokenizer(
        df_test["text"].tolist(), truncation=True, padding="max_length", max_length=512
    )
    test_dataset = NewsDataset(test_enc, df_test["label_id"].tolist())

    # Load the best model (from fold 1 for simplicity)
    model_dir = f"./final_model_{model_key}_fold1"
    model = AutoModelForSequenceClassification.from_pretrained(model_dir, num_labels=len(label_names)).to(device)
    model.eval()

    trainer = Trainer(model=model)
    preds = trainer.predict(test_dataset)

    y_true = preds.label_ids
    y_pred = np.argmax(preds.predictions, axis=1)

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    conf_matrices[model_key].append(cm)

    # Evaluation Metrics
    acc = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, average="weighted", zero_division=0)
    rec = recall_score(y_true, y_pred, average="weighted", zero_division=0)
    f1 = f1_score(y_true, y_pred, average="weighted", zero_division=0)
    macro_f1 = f1_score(y_true, y_pred, average="macro", zero_division=0)

    # Store Test Results
    results.append({
        "Model": model_key,
        "Fold": "Test",
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1": f1,
        "Macro_F1": macro_f1
    })

    # ✅ Collect Per-Class Metrics for Test Set
    report_dict = classification_report(y_true, y_pred, target_names=label_names, output_dict=True, zero_division=0)
    report_df = pd.DataFrame(report_dict).transpose()
    report_df["Model"] = model_key
    report_df["Fold"] = "Test"
    report_df.reset_index(inplace=True)
    report_df.rename(columns={"index": "Class"}, inplace=True)
    all_per_class_dfs.append(report_df)

    print(f"\n📊 Test Set Report ({model_key.upper()}):")
    print(classification_report(y_true, y_pred, target_names=label_names, zero_division=0))

    # Optional: Display Confusion Matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_names)
    disp.plot(xticks_rotation=45, cmap="Purples")
    plt.title(f"{model_key.upper()} - Test Set Confusion Matrix")
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# === ✅ Part 6: Word2Vec Training and Traditional Model Setup ===

# === ✅ Train Word2Vec Model ===
print("🔄 Training Word2Vec model...")
w2v_corpus = [simple_preprocess(text) for text in df_trainval["text"]]
w2v_model = Word2Vec(sentences=w2v_corpus, vector_size=100, window=5, min_count=2, workers=4, seed=42)
w2v_model.train(w2v_corpus, total_examples=len(w2v_corpus), epochs=10)

# === ✅ Embedding Vectorizer Using Mean Word Embeddings ===
class MeanEmbeddingVectorizer(BaseEstimator, TransformerMixin):
    def __init__(self, model):
        self.model = model
        self.vector_size = model.vector_size

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return np.array([
            np.mean([self.model.wv[w] for w in simple_preprocess(doc) if w in self.model.wv]
                    or [np.zeros(self.vector_size)], axis=0)
            for doc in X
        ])

# === ✅ Wrapper Class for MLPClassifier ===
class MLPClassifierWrapper(BaseEstimator):
    def __init__(self):
        self.model = MLPClassifier(hidden_layer_sizes=(100,), max_iter=200, random_state=42)

    def fit(self, X, y):
        self.model.fit(X, y)
        return self

    def predict(self, X):
        return self.model.predict(X)

    def predict_proba(self, X):
        return self.model.predict_proba(X)

# === ✅ Define Traditional Model Pipelines ===
traditional_models = {
    "tfidf_lr": Pipeline([
        ("tfidf", TfidfVectorizer(max_features=3000)),
        ("clf", LogisticRegression(max_iter=200, class_weight="balanced", random_state=42))
    ]),
    "word2vec_mlp": Pipeline([
        ("w2v", MeanEmbeddingVectorizer(w2v_model)),
        ("mlp", MLPClassifierWrapper())
    ])
}

print("✅ Traditional models initialized.")

# === ✅ Part 7: Train and Evaluate Traditional Models ===

for model_key, pipeline in traditional_models.items():
    print(f"\n🚀 Training Traditional Model: {model_key.upper()}")

    for fold, (train_idx, val_idx) in enumerate(skf.split(df_trainval["text"], df_trainval["label_id"])):
        print(f"📂 Fold {fold + 1}/5")

        X_train, X_val = df_trainval.iloc[train_idx]["text"], df_trainval.iloc[val_idx]["text"]
        y_train, y_val = df_trainval.iloc[train_idx]["label_id"], df_trainval.iloc[val_idx]["label_id"]

        # Train model
        pipeline.fit(X_train, y_train)
        y_pred = pipeline.predict(X_val)

        # Confusion Matrix
        cm = confusion_matrix(y_val, y_pred)
        conf_matrices.setdefault(model_key, []).append(cm)

        # Metrics
        acc = accuracy_score(y_val, y_pred)
        prec = precision_score(y_val, y_pred, average="weighted", zero_division=0)
        rec = recall_score(y_val, y_pred, average="weighted", zero_division=0)
        f1 = f1_score(y_val, y_pred, average="weighted", zero_division=0)
        macro_f1 = f1_score(y_val, y_pred, average="macro", zero_division=0)

        # Store Results
        results.append({
            "Model": model_key,
            "Fold": fold + 1,
            "Accuracy": acc,
            "Precision": prec,
            "Recall": rec,
            "F1": f1,
            "Macro_F1": macro_f1
        })

        # ✅ Collect Per-Class Metrics
        report_dict = classification_report(y_val, y_pred, target_names=label_names, output_dict=True, zero_division=0)
        report_df = pd.DataFrame(report_dict).transpose()
        report_df["Model"] = model_key
        report_df["Fold"] = fold + 1
        report_df.reset_index(inplace=True)
        report_df.rename(columns={"index": "Class"}, inplace=True)
        all_per_class_dfs.append(report_df)

        print(f"\n📊 Fold {fold + 1} Report ({model_key.upper()}):")
        print(classification_report(y_val, y_pred, target_names=label_names, zero_division=0))

    # === ✅ Evaluate on Hold-out Test Set ===
    print(f"\n🧪 Evaluating {model_key.upper()} on Hold-out Test Set")
    X_test = df_test["text"]
    y_test = df_test["label_id"]
    y_pred_test = pipeline.predict(X_test)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred_test)
    conf_matrices.setdefault(model_key, []).append(cm)

    # Metrics
    acc = accuracy_score(y_test, y_pred_test)
    prec = precision_score(y_test, y_pred_test, average="weighted", zero_division=0)
    rec = recall_score(y_test, y_pred_test, average="weighted", zero_division=0)
    f1 = f1_score(y_test, y_pred_test, average="weighted", zero_division=0)
    macro_f1 = f1_score(y_test, y_pred_test, average="macro", zero_division=0)

    results.append({
        "Model": model_key,
        "Fold": "Test",
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1": f1,
        "Macro_F1": macro_f1
    })

    # ✅ Collect Per-Class Metrics for Test
    report_dict = classification_report(y_test, y_pred_test, target_names=label_names, output_dict=True, zero_division=0)
    report_df = pd.DataFrame(report_dict).transpose()
    report_df["Model"] = model_key
    report_df["Fold"] = "Test"
    report_df.reset_index(inplace=True)
    report_df.rename(columns={"index": "Class"}, inplace=True)
    all_per_class_dfs.append(report_df)

    print(f"\n📊 Test Set Report ({model_key.upper()}):")
    print(classification_report(y_test, y_pred_test, target_names=label_names, zero_division=0))

    # Optional: Display Confusion Matrix
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=label_names)
    disp.plot(xticks_rotation=45, cmap="Purples")
    plt.title(f"{model_key.upper()} - Test Set Confusion Matrix")
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# === ✅ Part 8: Results Aggregation, Per-Class Metrics, Model Comparison Plots ===

# === ✅ Convert Results to DataFrame ===
results_df = pd.DataFrame(results)

# Split validation and test
val_results = results_df[results_df["Fold"] != "Test"]
test_results = results_df[results_df["Fold"] == "Test"]

# === ✅ Validation Summary (Averaged over folds) ===
val_summary = val_results.groupby("Model")[["Accuracy", "Precision", "Recall", "F1", "Macro_F1"]].mean().reset_index()
print("\n📊 Validation Performance Summary (Averaged Across Folds):")
print(val_summary.round(4).to_string(index=False))

# === ✅ Test Summary (Single evaluation per model) ===
print("\n🧪 Hold-out Test Set Performance Summary:")
print(test_results[["Model", "Accuracy", "Precision", "Recall", "F1", "Macro_F1"]].round(4).to_string(index=False))

# === ✅ Overall Confusion Matrices ===
print("\n📌 Overall Confusion Matrices Across All Folds:")
for model_key, cm_list in conf_matrices.items():
    overall_cm = sum(cm_list)
    disp = ConfusionMatrixDisplay(confusion_matrix=overall_cm, display_labels=label_names)
    disp.plot(xticks_rotation=45, cmap="Purples")
    plt.title(f"{model_key.upper()} - Overall Confusion Matrix (Folds + Test)")
    plt.grid(False)
    plt.tight_layout()
    plt.show()

# === ✅ Comparison Plot: F1 and Macro-F1 Scores ===
fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.35
models = test_results["Model"].tolist()
f1_scores = test_results["F1"].tolist()
macro_f1_scores = test_results["Macro_F1"].tolist()
x = np.arange(len(models))

ax.bar(x - bar_width/2, f1_scores, width=bar_width, label="F1 Score")
ax.bar(x + bar_width/2, macro_f1_scores, width=bar_width, label="Macro F1 Score")
ax.set_xticks(x)
ax.set_xticklabels(models, rotation=45)
ax.set_ylabel("Score")
ax.set_title("Model Comparison (F1 and Macro F1) on Test Set")
ax.legend()
plt.tight_layout()
plt.grid(axis='y')
plt.show()

# === ✅ Per-Class Metrics Aggregation ===
per_class_results_df = pd.concat(all_per_class_dfs, axis=0)

# Drop metrics like "accuracy", keep only per-class + macro/weighted
per_class_filtered = per_class_results_df[~per_class_results_df["Class"].isin(["accuracy"])]

# Average per class across folds and test
#overall_per_class_summary = per_class_filtered.groupby(["Model", "Class"]).mean().reset_index()
overall_per_class_summary = per_class_filtered.groupby(["Model", "Class"])[["precision", "recall", "f1-score", "support"]].mean(numeric_only=True).reset_index()

print("\n📋 Overall Per-Class Performance Summary (F1, Precision, Recall):")
print(overall_per_class_summary[["Model", "Class", "precision", "recall", "f1-score", "support"]].round(3).to_string(index=False))

# === ✅ Optional: Plot Per-Class F1, Precision, Recall
def plot_per_class_metric(df, metric="f1-score"):
    for model in df["Model"].unique():
        model_df = df[df["Model"] == model]
        class_metrics = model_df[~model_df["Class"].isin(["macro avg", "weighted avg", "accuracy"])]
        plt.figure(figsize=(8, 4))
        plt.bar(class_metrics["Class"], class_metrics[metric])
        plt.title(f"{model.upper()} - Per-Class {metric.title()}")
        plt.xticks(rotation=45)
        plt.ylabel(metric.title())
        plt.ylim(0, 1)
        plt.grid(axis="y")
        plt.tight_layout()
        plt.show()

# 🔧 Plot F1, Precision, and Recall per model
plot_per_class_metric(overall_per_class_summary, "f1-score")
plot_per_class_metric(overall_per_class_summary, "precision")
plot_per_class_metric(overall_per_class_summary, "recall")

# === ✅ Part 9: Statistical Significance Testing Between Model Categories ===

# Filter validation scores for statistical tests
transformer_keys = ["afriberta", "afroxlmr"]
traditional_keys = ["tfidf_lr", "word2vec_mlp"]

# Extract F1 and Macro-F1 from validation folds
val_f1 = val_results.pivot(index="Fold", columns="Model", values="F1")
val_macro_f1 = val_results.pivot(index="Fold", columns="Model", values="Macro_F1")

def run_significance_tests(metric_df, label="F1"):
    print(f"\n📊 Statistical Significance Tests ({label}):")
    for trans in transformer_keys:
        for trad in traditional_keys:
            if trans in metric_df.columns and trad in metric_df.columns:
                try:
                    t_stat, t_p = ttest_rel(metric_df[trans], metric_df[trad])
                    w_stat, w_p = wilcoxon(metric_df[trans], metric_df[trad])
                    print(f"{trans.upper()} vs {trad.upper()} -- Paired t-test p={t_p:.4f} | Wilcoxon p={w_p:.4f}")
                except Exception as e:
                    print(f"{trans.upper()} vs {trad.upper()} -- Error: {str(e)}")

# Run paired t-tests and Wilcoxon tests
run_significance_tests(val_f1, "F1")
run_significance_tests(val_macro_f1, "Macro-F1")

# === ✅ Part 10: Gradio GUI with Semantic Overlap and Explainability ===

# This file integrates conditional chunking into inference-time classification for Transformer models,
# while preserving all original Gradio GUI and semantic overlap explanation functionalities.
# Only additions are made; original logic remains untouched.

import gradio as gr
import difflib
import re
import torch
import os
import joblib
import pandas as pd
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from sklearn.metrics.pairwise import cosine_similarity

# === ✅ Highlight Overlap Function ===
def highlight_overlap(a, b):
    a_words = a.split()
    b_words = b.split()
    matcher = difflib.SequenceMatcher(None, a_words, b_words)
    a_highlighted, b_highlighted = [], []

    for tag, i1, i2, j1, j2 in matcher.get_opcodes():
        if tag == 'equal':
            a_highlighted.extend([f"<mark>{w}</mark>" for w in a_words[i1:i2]])
            b_highlighted.extend([f"<mark>{w}</mark>" for w in b_words[j1:j2]])
        else:
            a_highlighted.extend(a_words[i1:i2])
            b_highlighted.extend(b_words[j1:j2])

    return " ".join(a_highlighted), " ".join(b_highlighted)

# === ✅ Add: Chunking-Based Inference for Long Text Inputs ===
def classify_with_chunking(text, tokenizer, model, max_len=512, stride=256):
    tokens = tokenizer(text, return_tensors="pt", truncation=False)
    input_ids = tokens["input_ids"][0]
    token_count = len(input_ids)

    if token_count <= max_len:
        inputs = tokenizer(text, return_tensors="pt", truncation=True, padding='max_length', max_length=max_len).to(device)
        with torch.no_grad():
            logits = model(**inputs).logits[0]
        probs = torch.softmax(logits, dim=-1)
        return probs, torch.argmax(probs).item(), "Full input (≤512 tokens)", token_count

    else:
        chunks = []
        for i in range(0, len(input_ids), stride):
            chunk = input_ids[i:i + max_len]
            if len(chunk) < max_len:
                # ✅ Pad short final chunk to max_len
                pad_len = max_len - len(chunk)
                pad_token_id = tokenizer.pad_token_id if tokenizer.pad_token_id is not None else 0
                chunk = torch.cat([chunk, torch.tensor([pad_token_id] * pad_len)])
            chunks.append(chunk)

        logits_list = []
        for chunk in chunks:
            inputs = {"input_ids": torch.tensor([chunk]).to(device)}
            with torch.no_grad():
                logits = model(**inputs).logits[0]
            logits_list.append(logits)

        avg_logits = torch.stack(logits_list).mean(dim=0)
        probs = torch.softmax(avg_logits, dim=-1)
        return probs, torch.argmax(probs).item(), f"Chunked input: {len(chunks)} chunks", token_count

# === ✅ Update GUI Inference Function to Use Chunking Logic When Applicable ===
def classify_amharic_text_gui(text, model_choice="afriberta", top_k=3):
    try:
        if not text.strip():
            return "⚠️ Please enter valid Amharic text."

        text_clean = preprocess_amharic(text)
        label_map = joblib.load("label_map.pkl")
        input_vec = MeanEmbeddingVectorizer(w2v_model).transform([text_clean])
        all_vecs = MeanEmbeddingVectorizer(w2v_model).transform(df["text"].tolist())

        if model_choice in model_paths:
            model_dir = f"./final_model_{model_choice}_fold1"
            if not os.path.exists(model_dir):
                return f"❌ Model directory not found: {model_dir}"

            tokenizer = AutoTokenizer.from_pretrained(model_paths[model_choice])
            model = AutoModelForSequenceClassification.from_pretrained(model_dir).to(device)
            model.eval()

            # ✅ Use chunking-aware classifier
            probs, pred, path_used, token_count = classify_with_chunking(text_clean, tokenizer, model)
            scores = {label_map[i]: f"{p:.2f}" for i, p in enumerate(probs.cpu().numpy())}
            pred_label = label_map[pred]
            prob_text = "\n".join([f"{label}: {prob}" for label, prob in scores.items()])

        else:
            pipeline = traditional_models[model_choice]
            pred = pipeline.predict([text_clean])[0]
            probs = pipeline.predict_proba([text_clean])[0]
            pred_label = label_names[pred]
            scores = {label_names[i]: f"{p:.2f}" for i, p in enumerate(probs)}
            prob_text = "\n".join([f"{label}: {prob}" for label, prob in scores.items()])
            path_used = "Traditional model (no chunking)"
            token_count = len(text_clean.split())  # Approximate for traditional models

        sims = cosine_similarity(input_vec, all_vecs)[0]
        similar_indices = sims.argsort()[::-1]
        matches = []
        for i in similar_indices:
            if df.iloc[i]["label"] != pred_label:
                match_text = df.iloc[i]["text"]
                sim_score = sims[i]
                matches.append((match_text, df.iloc[i]["label"], sim_score))
            if len(matches) >= top_k:
                break

        display = f"🧠 **Predicted Class:** `{pred_label}`\n\n" \
                  f"📏 **Token Count:** `{token_count}`\n\n" \
                  f"📈 **Inference Path:** {path_used}\n\n" \
                  f"🔍 **Confidence Scores:**\n{prob_text}\n\n" \
                  f"📎 **Top-{top_k} Semantically Similar Texts (from other classes):**\n\n"

        for j, (sim_text, sim_label, sim_score) in enumerate(matches):
            input_high, sim_high = highlight_overlap(text_clean, sim_text)
            display += f"🔸 Match {j+1} (Class: `{sim_label}`, Similarity: {sim_score:.2f}`):\n\n"
            display += f"**Input Text:**\n{input_high}\n\n"
            display += f"**Similar Text:**\n{sim_high}\n\n"

        return display

    except Exception as e:
        return f"❌ Error during prediction: {str(e)}"

# === ✅ Gradio App Launch ===
gr.Interface(
    fn=classify_amharic_text_gui,
    inputs=[
        gr.Textbox(label="እባክዎ የአማርኛውን ዜና ያስገቡ / Please Enter Amharic News Here", lines=3),
        gr.Radio(label="Choose Model", choices=list(model_paths.keys()) + list(traditional_models.keys()), value="afriberta"),
        gr.Slider(minimum=1, maximum=5, step=1, value=3, label="Top-K Similar Texts")
    ],
    outputs=gr.Markdown(),
    title="የአማርኛ ዜና መለያ ስርዓት / Amharic News Classification System 📰",
    description="🧠 Predict news categories using Transformer and Traditional models.\n\n🔎 Includes semantic overlap explanations between your input and similar examples from other classes (Explainable AI).",
    allow_flagging="never"
).launch(share=True, debug=False)
```

```python
# === ✅ Part 4: Define API Clients for ChatGPT-4o, Gemini Pro, and Claude 3 ===

# --- ✅ ChatGPT-4o via OpenAI ---
import openai

def query_openai_gpt(text):
    try:
        prompt = f"""Classify the following Amharic news into one of the categories:
[accident, agriculture, economy, education, health, politics, sport, technology].

Text: \"{text}\""""

        response = openai.ChatCompletion.create(
            model="gpt-4o",
            api_key=os.getenv("OPENAI_API_KEY"),
            messages=[
                {"role": "system", "content": "You are a news classifier."},
                {"role": "user", "content": prompt}
            ],
            temperature=0.0
        )
        return response.choices[0].message["content"].strip().lower()
    except Exception as e:
        print(f"❌ OpenAI GPT-4o Error: {e}")
        return None

# --- ✅ Gemini Pro via Google GenerativeAI ---
import google.generativeai as genai

genai.configure(api_key=os.getenv("GEMINI_API_KEY"))
gemini_model = genai.GenerativeModel("gemini-pro")

def query_gemini(text):
    try:
        prompt = f"""Classify this Amharic news into one of these categories:
[accident, agriculture, economy, education, health, politics, sport, technology].

Text: \"{text}\""""

        response = gemini_model.generate_content(prompt)
        return response.text.strip().lower()
    except Exception as e:
        print(f"❌ Gemini Error: {e}")
        return None

# --- ✅ Claude 3 via Anthropic API ---
import anthropic

claude_client = anthropic.Anthropic(api_key=os.getenv("CLAUDE_API_KEY"))

def query_claude(text):
    try:
        prompt = f"""Classify the following Amharic news article into one of these categories:
[accident, agriculture, economy, education, health, politics, sport, technology].

Text: \"{text}\""""

        response = claude_client.messages.create(
            model="claude-3-opus-20240229",
            max_tokens=30,
            temperature=0,
            messages=[{"role": "user", "content": prompt}]
        )
        return response.content[0].text.strip().lower()
    except Exception as e:
        print(f"❌ Claude Error: {e}")
        return None
```

```python
# === ✅ Part 5: Run Predictions + Evaluate Accuracy, Precision, Recall, F1, Macro-F1 ===

from sklearn.metrics import classification_report

# Use the final hold-out test set
test_texts = test_df["clean_text"].tolist()
true_labels = test_df["label"].tolist()

# Store predictions from all LLMs
predictions = {
    "ChatGPT-4o": [],
    "Gemini": [],
    "Claude": []
}

# === ⚠️ May take time depending on sample size and rate limits
print("⚡ Running predictions on hold-out test set...")

for idx, text in enumerate(test_texts):
    print(f"⏳ Sample {idx+1}/{len(test_texts)}")

    # Query each model
    pred_gpt = query_openai_gpt(text)
    pred_gemini = query_gemini(text)
    pred_claude = query_claude(text)

    # Save predictions
    predictions["ChatGPT-4o"].append(pred_gpt)
    predictions["Gemini"].append(pred_gemini)
    predictions["Claude"].append(pred_claude)

# === ✅ Evaluation Function
from sklearn.metrics import accuracy_score, precision_recall_fscore_support

def evaluate_model(preds, name):
    print(f"\n📊 Evaluation for {name}:")
    preds = [p if p in label2id else "unknown" for p in preds]
    valid_idx = [i for i, p in enumerate(preds) if p != "unknown"]

    y_true = [true_labels[i] for i in valid_idx]
    y_pred = [preds[i] for i in valid_idx]

    acc = accuracy_score(y_true, y_pred)
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average="macro")

    print(classification_report(y_true, y_pred, zero_division=0))
    return {
        "model": name,
        "accuracy": acc,
        "precision": precision,
        "recall": recall,
        "f1": f1
    }

# Evaluate each model
results = []
for model_name, preds in predictions.items():
    result = evaluate_model(preds, model_name)
    results.append(result)
```
